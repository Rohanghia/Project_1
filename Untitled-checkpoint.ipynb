```python
#now let's analyze the home record and away record of Manchester United. We will divide the section into first and second leg of EPL
# first leg = first half of premier league - Aug - Dec
#second leg = second hald of premier league - Jan - May

allresult = pd.read_csv(match_results)
```

```python
#sort by date
allresult['Date'] = pd.to_datetime(allresult['Date'])

# define home and away for Manchester United
manutd_home = allresult[allresult.HomeTeam == 'Manchester United']
manutd_away = allresult[allresult.AwayTeam == 'Manchester United']

#define first half and Second half of the premier league for Manchester United

first_leg = manutd_home[manutd_home.Date < '2022-01-05'].append(manutd_away[manutd_away.Date < '2022-01-05']) 
second_leg = manutd_home[manutd_home.Date > '2022-01-05'].append(manutd_away[manutd_away.Date > '2022-01-05']) 
```

```python
# Let's Analyze all First Half results for Manchester United

first_leg.sort_values('Date', inplace=True)
first_leg.set_index('Date', inplace=True)

#code to analyze the Won, Draw and Lost at home and away

def wdl_home(data):
    
    decision = []
    gf = []
    ga = []
    home = data[data.HomeTeam == 'Manchester United']
    
    for u, v in [row.split(':') for row in home.Result.tolist()]:
        
        u = int(u)
        v = int(v)
        if u > v:
            decision.append('W')
            gf.append(u)
            ga.append(v)
        elif u < v:
            decision.append('L')
            gf.append(u)
            ga.append(v)
        else:
            decision.append('D')
            gf.append(u)
            ga.append(v)
    home['WDL'] = decision
    home['Scored'] = gf
    home['Conceded'] = ga
    home['GD'] = home['Scored'] - home['Conceded']
    return home

def wdl_away(data):
    
    decision = []
    gf = []
    ga = []
    
    away = data[data.AwayTeam == 'Manchester United']
    
    for u, v in [row.split(':') for row in away.Result.tolist()]:
        
        u = int(u)
        v = int(v)
        if u > v:
            decision.append('L')
            gf.append(v)
            ga.append(u)
        elif u < v:
            decision.append('W')
            gf.append(v)
            ga.append(u)
        else:
            decision.append('D')
            gf.append(v)
            ga.append(u)
            
    away['WDL'] = decision
    away['Scored'] = gf
    away['Conceded'] = ga
    away['GD'] = away['Scored'] - away['Conceded']
    return away
```

```python
#define first half for home and away
home_1stleg = wdl_home(first_leg)
away_1stleg = wdl_away(first_leg)
```

```python
#bar graph for the first leg
fig = plt.figure(figsize=(15, 5))
fig.subplots_adjust(wspace=0.3)
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
home_1stleg.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Manutd at home',
                                        color='red', ax=ax1);
ax1.set_ylabel('counts')
ax1.set_title('WDL home')
ax1.legend();
away_1stleg.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Manutd at away',
                                        color='blue', ax=ax2);
ax2.set_title('WDL away')
ax2.set_ylabel('counts')
ax2.legend();
```

```python
#average gaols scored and conceded during first half of the league

home_1stleg.mean()
```

```python
# let's analyze second half results for Manchester United

second_leg.sort_values('Date', inplace=True)
second_leg.set_index('Date', inplace=True)

#define second half for home and away
home_2ndleg = wdl_home(second_leg)
away_2ndleg = wdl_away(second_leg)
```

```python
# bar graph for the secong leg

fig1 = plt.figure(figsize=(15, 5))
fig1.subplots_adjust(wspace=0.3)
axis1 = fig1.add_subplot(1,2,1)
axis2 = fig1.add_subplot(1,2,2)
home_2ndleg.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Manutd at home',
                                        color='red', ax=axis1);
axis1.set_ylabel('counts')
axis1.set_title('WDL home')
axis1.legend();
away_2ndleg.WDL.value_counts().plot.bar(edgecolor='black', linewidth=1.2, label='Manutd at away',
                                        color='blue', ax=axis2);
axis2.set_title('WDL away')
axis2.set_ylabel('counts')
axis2.legend();
```

```python
# average gaols scored and conceded during second half of the league
home_2ndleg.mean()
```
